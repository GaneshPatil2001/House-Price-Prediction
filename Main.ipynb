#Standard Scaler
Why we use it?
Because we use the gradient descent to reach the global minima that's why we need to all input variables are on the same scale.

```python
## Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

# R square and adusted R square

## R² (Coefficient of Determination)

\[
R^2 = 1 - \frac{SSR}{SST}
\]

Where:

- **R²** = Coefficient of Determination  
- **SSR** = Sum of Squares of Residuals  
- **SST** = Total Sum of Squares  

```python
from sklearn.metrics import r2_score
score= r2_score(Y_test,reg_prdict)
print(score)
```

## Adjusted R²

\[
\text{Adjusted } R^2 = 1 - \left( \frac{(1 - R^2)(n - 1)}{n - k - 1} \right)
\]

Where:

- **R²** = Coefficient of Determination  
- **n** = Number of observations (samples)  
- **k** = Number of independent variables (features)  

```python
1-(1-score)*(len(Y_test)-X_test.shape[1]-1)
```
